클래스 메서드 => 클래스.메서드 / 클래스가 쓰는 메서드

인스턴스 메서드 : 인스턴스의 변수를 조작하거나 동작을 수행
ex) [1, 2, 3].append() => list클래스의 인스턴스 / append는 list클래스의 인스턴스 메서드

self : 인스턴스 자기 자신

=> 인스턴스 변수들의 초기값을 설정
※ 없어도 인스턴스 변수가 만들어 지긴함

클래스 메서드 => 클래스가 호출하는 메서드
=> 클래스 변수를 조작, 클래스 레벨의 동작 수행
클래스 변수 수정

스태틱(정적) 메서드 => 클래스와 인스턴스와 상관없이 독립적으로 동작하는 메서드
=> 클래스와 관련이 있지만
=> 필수적으로 작성해야 할 매개변수가 없음
=> 객체 상태나 클래스 상태를 수정할 수 없고 only 기능(행동)만을 위한 메서드
