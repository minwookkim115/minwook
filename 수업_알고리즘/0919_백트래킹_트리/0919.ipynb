백트래킹
- 대표저인 예시 : n*n queen
    - 퀸을 첫 번째 열(0, 0)에 뒀을 때
    - 해당 퀸이 공격할 수 있는 두 번째 열에서는 퀸을 두지 않는다.
    - 두번째 열에 뒀을 때도 퀸이 갈 수 있는 모든 곳에 퀸을 두지 않는다.
    - 점점 퀸을 놓을 수 있는 자리가 줄어들게 된다.
- 여러가지 선택지들이 존재하는 상황에서 한가지를 선택
- 선택이 이루어지면 새로운 선택지들의 집합이 생성
- 선택을 반복하면서 최종상태에 도달
- 올바른 선택을 계속하면 목표상태에 도달
- 갔다가 뒤로 되돌아 가니까 Back tracking 이다.
- ※ 핵심
    - 못가는 부분은 짤라버리니까 혁신적으로 확인할 부분이 줄어든다
    - 전체의 경우의 수 고려
    - 가능성 없는 경우의 수를 제거 == 가지치기
    - 백트래킹에서는 가지치기와 초기화를 해줘야 한다.
- 백트래킹을 사용할 때는 많은 경우의 수를 제거할 수 있는가를 따져봐야 한다.

백트래킹 예시
- 당첨 리프 노드 찾기
- 루트에서 갈 수 있는 노드 선택
- 꽝 노드까지 가면 다시 최근 선택으로 되돌아와서 시작
- 루트까지 돌아가고 선택지가 없다면 답이 없다

백트래킹 vs 깊이우선탐색(DFS)
- N! 가지의 경우의 수를 가진 문제에 대해 깊이 우선탐색 하면 못품
- 백트래킹을 이용하여 필요 없는 부분 가지치기
- ※ 최악의 경우에는 어쩔 수 없이 지수함수시간(N!)이 필요하니 불가능
- 깊이우선탐색은 일단 놓고 판단, 백트래킹은 일단 cut 함

트리
- 사이클이 없는 연결 그래프
    - 사이클 : 끝까지 갔을때 시작점으로 돌아오는 것
    - 연결 그래프 : 모든 꼭지점(노드)이 서로 갈 수 있다. / 끊어지면 다른 트리
- 이진 트리
    - 자녀 노드가 둘 이하인 트리
    0. 이진 트리 종류
        - 완전 이진 트리
            - 마지막 레벨을 제외한 모든 노드는 모두 꽉 차야 함(자식 노드가 2개)
            - 마지막 레벨 노드는 꽉 안차있어도 되는데 왼쪽 부터 채워져아 함
        - 포화 이진 트리
            - 모든 레벨이 꽉 차있는 것
        - 나머지 이진 트리
            - 자녀가 둘 이하기만 하면 됨
    1. 순회 방법(어떤 데이터를 먼저 볼 것인가)
        - 전위(부모 -> 좌 -> 우) => 부모에 좌로 갈 수 있는 데 까지 쭉 다음에 우
        - 중위(좌 -> 부모 -> 우)
        - 후위(좌 -> 우- > 부모)
    2. 트리 저장 방법(트리를 어떻게 저장 했는가)
