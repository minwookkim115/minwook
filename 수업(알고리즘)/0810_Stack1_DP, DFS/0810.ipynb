DP(동적계획, Dynamic Programming)
- 최적화 문제를 해결하는 알고리즘
- 풀이 방법
    - 입력 크기가 작은 부분들의 문제를 모두 해결한 후
    - 그 해들을 이용하여 보다 큰 크기의 부분 문제들을 해결
    - 최종적으로 원래 주어진 입력의 문제를 해결
    - ex) 파스칼 삼각형 1, 11, 1331, 14641

DFS(깊이 우선 탐색)
- 비선형구조는 모든 자료를 빠짐없이, 중복없이 검색하는 것이 중요
- DFS는 그 방법중 하나 / BFS(넓이 우선 탐색)도 있음
- 방법
    - 시작점에서 갈 수 있는데 까지 가다가
    - 마지막에 만난 갈림길에서 다른길로 가서 탐색
    - 반복
    - 마지막에 만난 갈림길을 꺼내야 하므로 **스택 사용** / 재귀도 사용

DFS 알고리즘
- 시작 점 v를 결정해서 방문
- 점 v에 인접한 정점 중에서
- 방문하지 않은 점 w가 있으면, 점 v를 스택에push하고 w로 ㄱㄱ
- 주변에 갈데 없으면 스택을 pop하고 받은 v로 가서 안간곳 ㄱㄱ
- 스택이 공백이 될 때까지 반복
- ※ 이동을 위한 명확한 규칙이 필요
