# 파이썬 용어

1. 표현식과 값

표현식 => ex) 1 + 2 / 5 * 21 - 4 / 'hello' / greeting = 'hi'

표현식 평가를 통해 값이 반환된다.

문장 : 실행 가능한 동작을 기술하는 코드 (조건문, 반복문, 함수)

2. 타입 => 값과 연산자

1 + 2 => 1=값(피연산자) / +(연산자)

값(피연산자)
Numeric Type(숫자) => int(정수), float(실수), complex(복소수)
Sequence(순서 있음) Types => list, tuple, range
Text Sequence(순서 있음) Type => srt(문자열)
Set Types(집합) => set
Mapping Types => dict
Boolean => True / False 판별

산술 연산자
+, -, *, /, //, %, ** 등

연산자 우선순위 (**, -(음수부호), * / // %, + - 순서) '=' 할당 연산자(낮음)

-2 ** 4 / -(2 ** 4) / (-2) ** 4
   -16  /    -16    /     16

3. 변수

변수 => 값을 담는 도구 / 값을 참조하는 이름 / 무엇인가 바라보고 있다.
할당문 => degrees = 36.5 => 변수 degrees에 값 36.5를 **할당**했다.

변수 규칙
- 영문 알파벳, 언더스코어(_), 숫자로 구성
- 숫자로 시작 불가
- 대소문자 구분
- 파이썬 내부 예약어는 사용 불가

**※ 메모리의 모든 위치에 고유 식별 주소가 존재한다.**

객체 => 타입을 갖는 메모리 주소 내 값 / 값이 들어있는 상자
- 파이썬은 모든 값이 객체다

number = 10
double = 2 * number
print(double) # 20

number = 5
print(double) # 20

WHY? => number에 10할당, double에 20할당 / number에 5가 할당 될 뿐 20이 할당된 double이 바뀌지는 않음


4. style guide(pep8)

코드의 일관성과 가독성 향상을 위한 규칙과 권장 사항

- 변수명은 무엇을 위한 변수인지 직관적인 이름이 필요
    - **is로 시작하는 변수나 함수는 True 냐 False 냐를 암묵적으로 의미**
    - 단수냐 복수냐 중요 => num = [1, 2, 3] X / numbers = [1, 2, 3] O
    - 대문자의 변수(상수) => 고정된 값 ex) SECONS_PER_MINUTE = 60 / HOURS = 24
- 공백 4칸(tab)을 사용하여 들여쓰기
    - ※ 파이썬은 중괄호로 구분하지 않음
- 한줄의 길이 79자, 길면 줄바꿈(\) 사용
- 문자와 밑줄(_)을 사용하여 함수, 변수, 속성 이름 작성 => snake_case

**※ pythontutor.com => 변수의 값들이 어떤게 지정돼있는지 볼수 있음**
※ 주석은 # or ctrl + /, 여러줄 주석은 """ """ 을 사용(길게 설명할 때)
- 주석의 목적 
    - 코드 특정부분 설명, 임시로 코드 비활성화
    - 코드를 이해하고 문서화
    - 코드의 의도나 동작을 설명
