컴퓨터 실력 + 도메인 전문성 + 수학 및 통계지식 => 데이터 사이언스

데이터 사이언스에서 정보추출 5단계
1 문제정의 : 구글 주식이 오를까?
2 데이터 수집 : 과거 구글 주가(숫자를 봐야함) + 정책관련 기사들(언어)
3 데이터 전처리 : 주가 데이터를 가공 / 오류가 있는지 검사
4 데이터 분석 : 필요한 정보를 추출
5 결과 해석 및 공유 : 그래프나 표를 통해(시각화) 세미나
※ 궁극적 목표는 문제를 해결

2 데이터 수집
- 웹 스크래핑(Web Scrapint) : 웹 페이지에서 데이터 추출
- 웹 크롤링(Web Crawling) : 웹 페이지를 자동탐색, 데이터 수집
- Open API 활용 : 공개된 API를 통해 수집
- 데이터 공유 플랫폼 활용 : 데이터를 공유하고 활용할 수 있는 온라인 플랫폼
    - 종류: 캐글, Data.world, 데이콘(Dacon), 공공데이터 포털 등
    - 캐글 : 데이터 분석 경진대회 플랫폼, 데이터셋 공유, 토론 가능

[참고] csv
몇 가지 필드를 (,)쉼표로 구분한 텍스트 데이터 및 텍스트 파일
window에서는 엑셀로 표현

3 데이터 정처리(정제)
- 불완전하거나 오류가 있는 데이터를 제거 => 데이터 품질 개선
- 중복 데이터 제거
- 분석하기 적절한 형식으로 데이터 변환
※ 데이터 전처리 및 분석에 사용되는 파이썬 패키지 ★★★★★★★ 중요 ★★★★★★★
- Numpy, Pandas, Matplotlib
    - Numpy
    - 다차원 배열을 쉽게 처리하고 효율적으로 사용할 수 있도록 지원하는 패키지
    - 장점: 데이터가 많은수록 Python 반복문에 비해 훨씬 빠름
    - 한계: 데이터에 레이블을 붙이거나 데이터 삭제 어려움 / 유연성 부족
    - Pandas
    - 엑셀을 다루듯 고성능의 데이터 구조를 만들 수 있음
    - Series(1차원 배열), DataFrame(2차원 배열)
    - Matplotlib
    - 데이터 시각화를 위해 사용
    - 그래프와 도표를 생성하고 데이터를 시각적으로 표현
