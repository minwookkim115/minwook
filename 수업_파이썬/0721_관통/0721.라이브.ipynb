오픈 API
- 외부에서 사용할 수 있도록 무료로 개방된 API
=> OpenWeatherMap API

오픈 API 특징 및 주의사항
- 악성 사용자가 100만 개의 계정으로 API에 요청을 보내는 상황
    - 너무 많으면 서버가 견디지 못함
- so 오픈 API는 API KEY를 활용해 사용자를 확인함
    - 사용자 인증 or 회원가입을 하면 서버에서 API KEY를 발급해줌
    - 서버에 요청할 때마다 API KEY를 함께 보내서 난 정상이야 증명

- 한명이 1초에 100만 개의 요청을 보내는 상황
    - 이것도 서버가 견디지 못해
- 오픈 API는 사용량이 제한되어 있다
    - 공식문서에서 사용량 제한 보고 혹시 넘으면 요금 청구될 수도 있다.

JSON
- JavaScript Object Notation / 자바스크립트 객체 표기법
- 텍스트 기반의 데이터 형식
- 특징
    - 데이터는 중괄호로 둘러싸인 키-값 쌍의 집합으로 표현 like dict
    - 키 = 문자열 / 값 = 다양한 데이터
    - 값은 쉼표로 구분

※ 파싱 => 데이터를 의미있는 구조로 분석하고 해석하는 과정
- 원하는 형태로 변환
